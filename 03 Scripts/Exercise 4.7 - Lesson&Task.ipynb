```python
#Import most recent orders_products_merged pickle 
df_opm = pd.read_pickle(r"C:\Users\repla\Documents\02 Data\Prepared Data\orders_products_merged.pkl")
```

```python
# create a subset of the data
df_opm = df_opm[:1000000]
```

```python
# defining the 'Price_label' function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
#use new 'price_label' function
df_opm['price_range'] = df_opm.apply(price_label, axis=1)
```

```python
#check values within the column
df_opm['price_range'].value_counts(dropna = False)
```

# If-statements with the loc() function

```python
#create new conditions
df_opm.loc[df_opm['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df_opm.loc[(df_opm['prices'] <= 15) & (df_opm['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
df_opm.loc[df_opm['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
#check value counts
df_opm['price_range_loc'].value_counts(dropna = False)
```

# If-statements with For-loops

```python
# find frequency of 'orders_day_of_week'
df_opm['orders_day_of_week'].value_counts(dropna = False)
```

```python
df_opm.head()
```

```python
#Create new column 'busiest day' using for-loop
result = []

for value in df_opm["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
#Print result
result
```

```python
#create new column in datafram
df_opm['busiest_day'] = result
```

```python
df_opm['busiest_day'].value_counts(dropna = False)
```

```python
#2 - Create new if-statement for-loop for new requirements
result_v2 = []

for value in df_opm["orders_day_of_week"]:
  if value == 0 or value == 1:
    result_v2.append("Busiest days")
  elif value == 4 or value == 3:
    result_v2.append("Least busy days")
  else:
    result_v2.append("Regularly busy")
```

```python
#4 - Check frequency of values in the "time_of_order" column 
df_opm['time_of_order'].value_counts()
```

```python
# Create a for-loop if statement 
hour_of_day = []

for value in df_opm["time_of_order"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    hour_of_day.append("Most orders")
  elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
    hour_of_day.append("Fewest orders")
  else:
    hour_of_day.append("Average orders")
```

```python
# Add result to dataframe as "busiest_period_of_day" column
df_opm['busiest_period_of_day'] = hour_of_day
```
